### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv.

### Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”.

```python
produrl =[]
```

```python
produrl_tag = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
```

```python
for u in produrl_tag:
 try: 
     produrl.append(u.get_attribute("href"))
 except:
    produrl.append('-')
```

```python
import time 


i=2

while(i>0): 
    nextbtn = driver.find_element_by_xpath("//li[@class='a-last']")
    nextbtn.click()
    time.sleep(3)
    produrl_tag = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for u in produrl_tag:
          produrl.append(u.get_attribute("href"))
    i=i-1
```

```python
Brand = []
prod = []
rate = []
ratenum = []
price = []
ex = []
delv = []
avl = []
det = []
```

```python
for i in produrl:
    driver.get(i)
    time.sleep(5)
    try:
        brandtag = driver.find_element_by_xpath("//td[@class='a-size-base prodDetAttrValue']")
        Brand.append(brandtag.text)
    except:
        Brand.append('-')
    try:
        prodtag = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        prod.append(prodtag.text)
    except:
        prod.append('-')
    try:
        ratetag = driver.find_element_by_xpath("//div[@id='averageCustomerReviews']//span[@class='a-icon-alt']")
        rate.append(ratetag.get_attribute("innerHTML"))
    except NoSuchElementException:
        rate.append('-')
    try:
        ratenumtag = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        ratenum.append(ratenumtag.get_attribute("innerHTML"))
    except:
        ratenum.append('-')
    try:
        pricetag = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        price.append(pricetag.text)
    except:
        price.append('-')
    
    try:
        extag = driver.find_element_by_xpath("//div[@class='a-accordion-row-a11y']//h5")
        ex.append(extag.text.replace('\n',' ').replace('   ',''))
    except:
        ex.append('-')
    
    try:
        delvtag = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        delv.append(delvtag.text)
    except:
        delv.append('-')
    try:
        avltag = driver.find_element_by_xpath("//div[@id='availability']//span")
        avl.append(avltag.text)
    except:
        avl.append('-')
    try:
        dettag = driver.find_element_by_xpath("//div[@id='feature-bullets']")
        det.append(dettag.text.replace('\nShow More','.').replace('\n',''))
    except:
        det.append('-')
```

```python
Laptops = {'Brand Name':Brand,'Product Name':prod,'Price(₹)':price,'Rating':rate,'Number of Ratings':ratenum ,'Exchange':ex,'Expected Delivery':delv,'Availablitiy':avl,'Details':det,'Product URL':produrl}

Laptops  = pd.DataFrame(Laptops)
```

### 3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
a = ['fruits','cars','Machine Learning']
```

```python
for k in a:
    url =  'https://www.images.google.com/'
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)
    search = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/form/div[1]/div[1]/div[1]/div/div[2]/input')
    search.send_keys(k)
    time.sleep(3)
    searchbtn = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/form/div[1]/div[1]/div[1]/button')
    searchbtn.click()
    time.sleep(5)
    imgsrc = []
    imgurl = []
    image_tag = driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    i=0
    while(i<320):
        driver.execute_script("window.scrollBy(0,500000)") # scroll down to get more images
        i+=1
    time.sleep(5)
    image_tag = driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    for i in image_tag:
        imgurl.append(i.get_attribute('src'))
    for i in imgurl:
        if i != None and i[0:8] =='https://': #extracing image urls
             imgsrc.append(i)
        
    for i in range(0,100):
       d = requests.get(imgsrc[i])
       with open("E:/imgs/"+k+str(i)+".png","wb") as file: #opening as file to be written in binary mode
        
          file.write(d.content) #saving image in direcotry from source url
    driver.quit()     
```
